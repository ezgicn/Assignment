```python
import statsmodels.api as sm
import numpy as np
import patsy
from patsy import dmatrices

y, X = patsy.dmatrices('y13 ~ x', data=df, return_type="dataframe")
model = sm.OLS(y, X)
fit = model.fit()
ypred = fit.predict(X)

#Mean of Residuals for y13
residuals = df["y13"]-ypred 
mean_residuals = np.mean(residuals)
print(f"Mean of Residuals: {mean_residuals:.10f}")

#Homoscedasticity for y13
from scipy.stats import bartlett
stat, p = bartlett(df['x'],residuals)
print("Bartlett Test p-value: {:g}".format(p))

#Normality of error terms for y13
from scipy import stats
stat, p = stats.normaltest(residuals)
print(f"p-value normality: {p:.10f}")

#Autocorrelation Test for y13
dw = sm.stats.durbin_watson(residuals)
print("Durbin-Watson:",dw)
```
